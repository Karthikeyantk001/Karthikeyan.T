```python
import matplotlib.pyplot as plt
```

```python
plt
```

```python
years=[x for x in range(2000,2010)]
```

```python
pops=[1.2,2.3,3,4.3,5,6,7,8,8.9,10.01]
```

```python
plt.plot(years,pops)
plt.xlabel("years")
plt.ylabel("population")
plt.title("keyan population data base ")
```

```python
deaths=[1.1,2.2,3.4,4.1,5.9,6.1,4.2,6.6,7.5,9.0]
plt.plot(years,pops)
plt.plot(years,deaths)
plt.xlabel("years")
plt.ylabel("population")
plt.title("keyan population data base ")
```

```python
plt.plot(years,pops)
plt.plot(years,deaths,"--")
plt.xlabel("years")
plt.ylabel("population")
plt.title("keyan population data base ")
```

```python
plt.plot(years,pops)
plt.plot(years,deaths,"--")
plt.xlabel("years")
plt.ylabel("population")
plt.grid(True)
plt.title("keyan population data base ")
```
